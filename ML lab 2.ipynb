```
import pandas as pd
import numpy as np

data = {
    'ID': [1, 2, 3, 4, 5],
    'Age': [25, 30, np.nan, 35, 40],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'Salary': [50000, 60000, 58000, np.nan, 62000],
    'Purchased': ['Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

df['Age'].fillna(df['Age'].mean(), inplace=True)

df['Salary'].fillna(df['Salary'].median(), inplace=True)
print(df,"\n")
```

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

df1 = pd.DataFrame({
    'OrderID': [101, 102, 103, 104],
    'Product': ['A', 'B', 'A', 'C'],
    'Quantity': [10, 5, 8, 12],
    'Price': [100, 200, 100, 300]
})

df2 = pd.DataFrame({
    'OrderID': [101, 102, 103, 105],
    'Customer': ['John', 'Anna', 'Mike', 'Sara'],
    'Age': [28, np.nan, 35, 40]
})

print("Original df1:\n", df1)
print("\nOriginal df2:\n", df2)
```

```
df1['TotalSale'] = df1['Quantity'] * df1['Price']
pivot_df = df1.pivot_table(values='TotalSale', index='Product', aggfunc='sum')
print("\nPivoted df1 (Total sales by Product):\n", pivot_df)
```

```
filtered_df = df1[df1['Quantity'] > 8]
print("\nFiltered df1 (Quantity > 8):\n", filtered_df)

merged_df = pd.merge(df1, df2, on='OrderID', how='inner')
print("\nMerged DataFrame (inner join on OrderID):\n", merged_df)

print("\nMissing values in 'Age':\n", merged_df['Age'].isna())

mean_age = merged_df['Age'].mean()
merged_df['Age'].fillna(mean_age, inplace=True)
print("\nDataFrame after filling missing 'Age' with mean:\n", merged_df)
```

```

scaler = MinMaxScaler()

cols_to_normalize = ['Quantity', 'Price', 'Age']
merged_df[cols_to_normalize] = scaler.fit_transform(merged_df[cols_to_normalize])

print("\nDataFrame after Min-Max Normalization:\n", merged_df)
```

```
import pandas as pd

data = {
    'id': range(1, 11),
    'first': ['Leone', 'Romola', 'Geri', 'Sandy', 'Jacenta', 'Diane-marie', 'Austen', 'Vanya', 'Giordano', 'Rozele'],
    'last': ['Debrick', 'Phinnessy', 'Prium', 'Doveston', 'Jansik', 'Medhurst', 'Pool', 'Teffrey', 'Elloy', 'Fawcett'],
    'gender': ['Female', 'Female', 'Male', 'Female', 'Female', 'Female', 'Male', 'Male', 'Male', 'Female'],
    'Marks': [50, 60, 65, 95, 31, 45, 45, 70, 36, 50],
    'selected': [True, False, False, False, True, True, True, False, False, False]
}

df = pd.DataFrame(data)
df.to_csv('sample.csv', index=False)
```

```
import pandas as pd

df = pd.read_csv('sample.csv')

print("DataFrame Head:\n", df.head(), "\n")

print("Mean Marks:", df['Marks'].mean())

print("Median Marks:", df['Marks'].median())

print("Mode Marks:", df['Marks'].mode().values)

print("Std Deviation Marks:", df['Marks'].std())

print("Variance Marks:", df['Marks'].var())

print("Min Marks:", df['Marks'].min())
print("Max Marks:", df['Marks'].max())

print("Count Selected:", df['selected'].sum())

print("Unique Genders:", df['gender'].nunique())

print("\nFrequency count for gender:\n", df['gender'].value_counts())

df['selected_int'] = df['selected'].astype(int)
print("Correlation between Marks and selected:", df['Marks'].corr(df['selected_int']))
```

```
print("\nDescriptive statistics for numeric columns:\n", df.describe())

print("\nDescriptive statistics for categorical columns:\n", df.describe(include=['object', 'bool']))
```

```
print("\nUnivariate Statistics for Marks:")
print("Mean:", df['Marks'].mean())
print("Median:", df['Marks'].median())
print("Mode:", df['Marks'].mode().values)
print("Standard Deviation:", df['Marks'].std())
print("Variance:", df['Marks'].var())

print("\nFrequency distribution of gender:")
print(df['gender'].value_counts())

print("\nCross-tabulation between gender and selected:")
print(pd.crosstab(df['gender'], df['selected']))

print("\nMean Marks by gender:")
print(df.groupby('gender')['Marks'].mean())

selected_pct = df.groupby('gender')['selected_int'].mean() * 100
print("\nPercentage of selected students by gender:")
print(selected_pct)
```

```
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler, Binarizer

df = pd.read_csv('sample.csv')

print("Original Data:")
print(df)

le_gender = LabelEncoder()
df['gender_encoded'] = le_gender.fit_transform(df['gender'])

df['selected_encoded'] = df['selected'].astype(int)

print("\nAfter Label Encoding:")
print(df[['gender', 'gender_encoded', 'selected', 'selected_encoded']])

scaler = StandardScaler()
df['Marks_scaled'] = scaler.fit_transform(df[['Marks']])

minmax_scaler = MinMaxScaler()
df['Marks_minmax_scaled'] = minmax_scaler.fit_transform(df[['Marks']])

print("\nAfter Scaling 'Marks':")
print(df[['Marks', 'Marks_scaled', 'Marks_minmax_scaled']])

binarizer = Binarizer(threshold=50)
df['Marks_binarized'] = binarizer.fit_transform(df[['Marks']])

print("\nAfter Binarizing 'Marks' with threshold=50:")
print(df[['Marks', 'Marks_binarized']])
```
