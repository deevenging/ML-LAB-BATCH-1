```
import math
import statistics
import numpy as np
from scipy import linalg
```

```
x=7.5

print("Floor:",math.floor(x))
print("Ceil:",math.ceil(x))
print("sqrt:",math.sqrt(16))
print("isqrt:",math.isqrt(16))
print("GCD:",math.gcd(60,48))
```

```
arr=np.array([[1,2,3],[4,5,6]])

print("ndim:",arr.ndim)
print("shape:",arr.shape)
print("size:",arr.size)

print("sum:",arr.sum())
print("mean:",arr.mean())
print("sort:",np.sort(arr,axis=None))
print("sin:",np.sin(arr))
```

```
A=np.array([[1,2],[2,5]])

det=linalg.det(A)
print("det:",det)

eigvals, eigvecs = linalg.eig(A)
print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)
```

```
lst = [1, 2, 3, 4, 5, 6]

arr1d = np.array(lst)
print("1D Array:\n",arr1d)

arr2d = arr1d.reshape(2, 3)
print("2D Array:\n", arr2d)

arr3d = arr1d.reshape(2, 1, 3)
print("3D Array:\n", arr3d)
```

```
zeros = np.zeros((2,2))
ones = np.ones((3,3))
rand = np.random.rand(2,3)

print("Zeros:\n", zeros)
print("Ones:\n", ones)
print("Random:\n", rand)
```

```
B = np.array([[5, 6], [7, 8]])
det_B = linalg.det(B)
print("Determinant of B:", det_B)
```

```
eigvals_B, eigvecs_B = linalg.eig(B)
print("Eigenvalues of B:\n", eigvals_B)
print("Eigenvectors of B:\n", eigvecs_B)
```

```
import pandas as pd

data_dict = {
    'Name': ['Abi', 'Sam', 'Jess'],
    'Age': [21, 22, 23],
    'Marks': [85, 90, 95]
}

df = pd.DataFrame(data_dict)
print("DataFrame from dict:\n", df)

data_list = [
    ['Abi', 21, 85],
    ['Sam', 22, 90],
    ['Jess', 23, 95]
]

df2 = pd.DataFrame(data_list, columns=['Name', 'Age', 'Marks'])
print("\nDataFrame from list:\n", df2)
```

```
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()

print("Iris keys:", iris.keys())
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

print("\nIris DataFrame:\n", df_iris.head())

from sklearn.datasets import load_diabetes

diabetes = load_diabetes()
df_diabetes = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df_diabetes['target'] = diabetes.target

print("\nDiabetes DataFrame:\n", df_diabetes.head())
```
